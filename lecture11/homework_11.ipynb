## Problem 1 - Calculating expectations for discrete random variables

Consider the Categorical random variable $X \sim \text{Categorical}(0.2, 0.4, 0.4)$ taking three discrete values $1, 2,$ and $3$.
Find the numerical answer for the following.
Hint: You can do it by hand or write some code. It is up to you.

+ $\mathbf{E}[X] = ?$

**Answer:**
<br><br>E[X] = 1×0.2 + 2×0.4 + 3×0.4 = 0.2 + 0.8 + 1.2 = 2.2

````python
import numpy as np
probs = np.array([0.2, 0.4, 0.4])
values = np.array([1, 2, 3])
EX = np.sum(values * probs)
print(f'E[X] = {EX}')
````

+ $\mathbf{E}[X^2] = ?$

**Answer:**
<br><br>E[X²] = 1²×0.2 + 2²×0.4 + 3²×0.4 = 0.2 + 1.6 + 3.6 = 5.4

````python
EX2 = np.sum(values**2 * probs)
print(f'E[X²] = {EX2}')
````

+ $\mathbf{V}[X] = ?$

**Answer:**
<br><br>V[X] = E[X²] - (E[X])² = 5.4 - 2.2² = 5.4 - 4.84 = 0.56

````python
VX = EX2 - EX**2
print(f'V[X] = {VX}')
````

+ $\mathbf{E}[e^X] = ?$

**Answer:**
<br><br>E[eˣ] = e¹×0.2 + e²×0.4 + e³×0.4 = 0.2e + 0.4e² + 0.4e³ ≈ 0.5436 + 2.9556 + 8.0428 ≈ 11.542

````python
E_exp_X = np.sum(np.exp(values) * probs)
print(f'E[e^X] = {E_exp_X:.4f}')
````

## Problem 2 - Calculate the expectation and variance of a continuous random variable

Take an exponential random variable, see [Problem 2 of Homework 10](lecture10:homework:problem2),
with rate parameter $\lambda$:

$$
T \sim \text{Exp}(\lambda).
$$

The PDF of the exponential is:

$$
p(t) = \lambda e^{-\lambda t},
$$

for $t\ge 0$ and zero otherwise.

+ Prove mathematically that $\mathbf{E}[T] = \lambda^{-1}$. Hint: You need to do the integral $\int_0^\infty tp(t)dt$ using [integration by parts](https://en.wikipedia.org/wiki/Integration_by_parts).

**Answer:**
<br><br>E[T] = ∫₀^∞ t·λe^(-λt) dt
Using integration by parts with u=t, dv=λe^(-λt)dt:
du=dt, v=-e^(-λt)
= [-te^(-λt)]₀^∞ + ∫₀^∞ e^(-λt) dt
The first term → 0 (since te^(-λt) → 0 as t→∞), so:
= [-e^(-λt)/λ]₀^∞ = 0 - (-1/λ) = 1/λ = λ⁻¹

+ Prove mathematically that $\mathbf{V}[T] = \lambda^{-2}$. Hint: Use integration by parts to find $\mathbf{E}[T^2]=\int_0^\infty t^2 p(t)dt$ and then use one of the properties of the variance.

**Answer:**
<br><br>First find E[T²] = ∫₀^∞ t²·λe^(-λt) dt
Integration by parts with u=t², dv=λe^(-λt)dt:
du=2t dt, v=-e^(-λt)
= [-t²e^(-λt)]₀^∞ + ∫₀^∞ 2t·e^(-λt) dt
= 0 + (2/λ)∫₀^∞ t·λe^(-λt) dt = (2/λ)·E[T] = (2/λ)·(1/λ) = 2/λ²
Then using V[T] = E[T²] - (E[T])²:
V[T] = 2/λ² - (1/λ)² = 2/λ² - 1/λ² = 1/λ² = λ⁻²

## Problem 3 - Properties of expectations

Let $X$ be a random variable with expectation $\mathbf{E}[X] = 3$ and variance $\mathbf{V}[X] = 2$.
Calculate the following expressions.

+ $\mathbf{E}[4X] = ?$

**Answer:**
<br><br>E[4X] = 4·E[X] = 4×3 = 12

+ $\mathbf{E}[X + 2] = ?$

**Answer:**
<br><br>E[X+2] = E[X] + 2 = 3 + 2 = 5

+ $\mathbf{E}[3X + 1] = ?$

**Answer:**
<br><br>E[3X+1] = 3·E[X] + 1 = 9 + 1 = 10

+ $\mathbf{V}[5X] = ?$

**Answer:**
<br><br>V[5X] = 5²·V[X] = 25×2 = 50

+ $\mathbf{V}[X + 3] = ?$

**Answer:**
<br><br>V[X+3] = V[X] = 2 (adding a constant doesn't change variance)

+ $\mathbf{V}[2X + 1] = ?$

**Answer:**
<br><br>V[2X+1] = 2²·V[X] = 4×2 = 8

+ $\mathbf{E}[X^2] = ?$

**Answer:**
<br><br>Using V[X] = E[X²] - (E[X])²:
E[X²] = V[X] + (E[X])² = 2 + 9 = 11
