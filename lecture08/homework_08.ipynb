## Problem 1 - Medical test analysis

We are tasked with assessing the usefulness of a tuberculosis test.
The prior information I is:

> The percentage of the population infected by tuberculosis is 0.4\%.
We have run several experiments and determined that:
+ If a tested patient has the disease, then 80\% of the time the test comes out positive.
+ If a tested patient does not have the disease, then 90\% of the time the test comes out negative.

To facilitate your analysis, consider the following logical sentences concerning a patient:

> A: The patient is tested and the test is positive.

> B: The patient has tuberculosis.

A. Find the probability that the patient has tuberculosis (before looking at the result of the test), i.e., $p(B|I)$. This is known as the base rate or the prior probability.

**Answer:**
<br><br>The percentage of the population infected with tuberculosis is directly given as 0.4%.
p(B|I) = 0.004

B. Find the probability that the test is positive given that the patient has tuberculosis, i.e., $p(A|B,I)$.

**Answer:**
<br><br>Given directly from the experiment: if the patient has the disease, the test is positive 80% of the time.
p(A|B,I) = 0.80

C. Find the probability that the test is positive given that the patient does not have tuberculosis, i.e., $p(A|\neg B, I)$.

**Answer:**
<br><br>If the patient does not have the disease, the test is negative 90% of the time, so it is positive the remaining 10%.
p(A|¬B,I) = 0.10

D. Find the probability that a patient that tested positive has tuberculosis, i.e., $p(B|A,I)$.

**Answer:**
<br><br>p(B|A,I) = p(A|B,I) × p(B|I) / p(A|I)
First find p(A|I) (total probability of a positive test):
p(A|I) = p(A|B,I)×p(B|I) + p(A|¬B,I)×p(¬B|I)
p(A|I) = 0.80×0.004 + 0.10×0.996 = 0.0032 + 0.0996 = 0.1028
Then:
p(B|A,I) = (0.80 × 0.004) / 0.1028 = 0.0032 / 0.1028 ≈ 0.0311 (about 3.1%)
This is surprisingly low! Even with a positive test, there is only a ~3.1% chance the patient actually has tuberculosis. This is due to the very low base rate (0.4%).

E. Find the probability that a patient that tested negative has tuberculosis, i.e., $p(B|\neg A, I)$. Does the test change our prior state of knowledge about about the patient? Is the test useful?

**Answer:**
<br><br>p(¬A|I) = 1 - 0.1028 = 0.8972
p(¬A|B,I) = 1 - 0.80 = 0.20 (false negative rate)
p(B|¬A,I) = p(¬A|B,I) × p(B|I) / p(¬A|I)
= (0.20 × 0.004) / 0.8972 = 0.0008 / 0.8972 ≈ 0.00089 (about 0.089%)
The prior was 0.4% and after a negative test it drops to 0.089%, so the test does update our knowledge. However, even a positive result only gives ~3.1% probability of disease, so the test is not very useful in practice for such a rare disease — most positive results are false positives.

F. What would a good test look like? Find values for

$$
p(A|B,I)= p(\text{test is positive} |\text{has tuberculosis},I),
$$

and

$$
p(A| \neg B, I) = p(\text{test is positive}|\text{does not have tuberculosis}, I),
$$

so that

$$
p(B|A, I) = p(\text{has tuberculosis}|\text{test is positive}, I) = 0.99.
$$

There are more than one solutions. How would you pick a good one? Thinking in this way can help you set goals if you work in R\&D. If you have time, try to figure out whether or not there exists such an accurate test for tuberculosis

**Answer:**
<br><br>We want: p(B|A,I) = 0.99, which means:
p(A|B,I) × 0.004 / [p(A|B,I)×0.004 + p(A|¬B,I)×0.996] = 0.99
Solving for the relationship:
p(A|B,I) × 0.004 = 0.99 × [p(A|B,I)×0.004 + p(A|¬B,I)×0.996]
One solution: set p(A|¬B,I) very small. For example:

p(A|B,I) = 0.99 (sensitivity), p(A|¬B,I) ≈ 0.000040 (extremely specific)

This means the false positive rate would need to be just 0.004% — an extremely specific test. In practice, a good test would maximize both sensitivity p(A|B,I) (close to 1) and minimize the false positive rate p(A|¬B,I) (close to 0). When the disease is very rare, specificity matters far more than sensitivity.
For tuberculosis specifically, modern tests like IGRA (Interferon-Gamma Release Assays) have high specificity (~95–99%), but achieving p(B|A,I) = 0.99 for a disease with only 0.4% prevalence would still require an almost perfect false-positive rate, which is extremely difficult to achieve in practice.
